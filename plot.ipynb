```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import logging
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import quad

import utils
from core import Plane3d, Ellipsoid
```

```python
plane = Plane3d(1, 2, 3, 1)
x = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x, x)
X = X.reshape((-1, 1))
Y = Y.reshape((-1, 1))
Z = plane.xy2z(np.hstack((X, Y)))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

```python
n_pts = 22
elps = Ellipsoid(5, 10, 5)
x = np.linspace(-4, 4, n_pts)
y = np.linspace(-8, 8, n_pts)
X, Y = np.meshgrid(x, y)
X = X.reshape((-1, 1))
Y = Y.reshape((-1, 1))
pt_xy = np.hstack((X, Y))
Z = elps.xy2z(pt_xy)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z)
ax.scatter(X, Y, -Z)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

```python
plane = Plane3d(1, 2, 3, 1)
elps = Ellipsoid(5, 10, 5)
rx, ry, rz = elps.rx, elps.ry, elps.rz
a, b, c, d = plane.a, plane.b, plane.c, plane.d

def arc_length(start_x, end_x):
    rs = np.array([elps.rx, elps.ry, elps.rz])
    return quad(lambda x: 2*np.linalg.norm(elps.plane_intersect_arc(plane, [x]).squeeze() / rs), start_x, end_x)

print arc_length(-4, 4)
```
